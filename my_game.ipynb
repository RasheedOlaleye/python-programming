```python
# import random
import random
```

```python
# Cell 2 — Word list
words = ["python", "hangman", "jupyter", "function", "variable"]
```

```python
def get_random_word(words):
    return random.choice(words)

def display_word(word, guessed_letters):
    return " ".join([letter if letter in guessed_letters else "_" for letter in word])

def is_won(word, guessed_letters):
    return all(letter in guessed_letters for letter in word)
```

```python
def play_hangman():
    word         = get_random_word(words)
    guessed      = []
    attempts     = 6
    
    print("🎮 Welcome to Hangman!")
    print(f"The word has {len(word)} letters\n")

    while attempts > 0:
        
        # Show current state
        print(display_word(word, guessed))
        print(f"Attempts left : {attempts}")
        print(f"Guessed so far: {guessed}\n")

        # Get input
        guess = input("Guess a letter: ").lower()

        # Validate input
        if len(guess) != 1 or not guess.isalpha():
            print("⚠️  Please enter a single letter!\n")
            continue

        if guess in guessed:
            print("⚠️  You already guessed that letter!\n")
            continue

        # Add to guessed list
        guessed.append(guess)

        # Check if correct
        if guess in word:
            print(f"✅ Good guess! '{guess}' is in the word!\n")

            # Check win
            if is_won(word, guessed):
                print(display_word(word, guessed))
                print(f"🎉 You win! The word was: '{word}'")
                return
        else:
            attempts -= 1
            print(f"❌ Wrong! '{guess}' is not in the word!\n")

            # Check lose
            if attempts == 0:
                print(f"💀 Game over! The word was: '{word}'")
                return
```

```python
play_hangman()
```
